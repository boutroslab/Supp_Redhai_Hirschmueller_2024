```python
# Create a bed file with relative window around genes
bed = pd.read_csv("../resources/genes/genes.bed", sep="\t", header=None)
bed.columns = ["chrom", "start", "end", "name"]
window_scale = 1/3  # gene length * window_scale
window = ((bed['end'] - bed['start']) * window_scale).astype(int)
bed['start'] = bed['start'] - window
bed['end'] = bed['end'] + window
bed.to_csv(f"../resources/genes/genes_relative_window.bed", sep="\t", header=None, index=None)
```
